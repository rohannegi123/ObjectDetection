```python

import cv2
import numpy as np
cfg = "C:\\Users\\rohan\\Downloads\\yolov3-tiny.cfg"
weights = "C:\\Users\\rohan\\Downloads\\yolov3-tiny.weights" 
coco_names = "C:\\Users\\rohan\\Downloads\\coco.names"

##loading packages and files
```

```python


net = cv2.dnn.readNet(weights , cfg )
classes = []
with open(coco_names, "r") as f:
    classes = [line.strip() for line in f.readlines()]
layer_names = net.getLayerNames()
output_layers = [layer_names[i[0] - 1] for i in net.getUnconnectedOutLayers()]
colors = np.random.uniform(0, 255, size=(len(classes), 3))



video = cv2.VideoCapture("C:\\Users\\rohan\\Downloads\\test.mp4")
#loading video with opencv
while True:
    _,frame = video.read()
    img = cv2.resize(frame, (700, 540)) 
    height, width, channels = img.shape

    # Detecting objects
    blob = cv2.dnn.blobFromImage(img, 0.00392, (320, 320), (0, 0, 0), True, crop=False)
    net.setInput(blob)
    outs = net.forward(output_layers)

    # points of polygon
    pts = np.array([[25, 70], [25, 160], 
                    [110, 200], [200, 160], 
                    [200, 70], [110, 20]],
                   np.int32) 

    pts = pts.reshape((-1, 1, 2))
    color = (255, 0, 0)
    thickness = 2
    image = cv2.polylines(img, [pts], True, color, thickness)    #polygon created      
    class_ids = []
    confidences = []
    boxes = []
    for out in outs: ##detection of objects
        for detection in out:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]
            if confidence > 0.5:
                # Object detected
                center_x = int(detection[0] * width)
                center_y = int(detection[1] * height)
                w = int(detection[2] * width * 2 )
                h = int(detection[3] * height * 2)
                # Rectangle coordinates
                x = int(center_x - w/5 )
                y = int(center_y - h/5 )
                boxes.append([x, y, w, h])
                confidences.append(float(confidence))
                class_ids.append(class_id)

    indexes = cv2.dnn.NMSBoxes(boxes, confidences, 0.5, 0.4)
 
    font = cv2.FONT_HERSHEY_PLAIN
    for i in range(len(boxes)):  ##creating frame over objects
        if i in indexes:
            x, y, w, h = boxes[i]
            label = str(classes[class_ids[i]])
            color = colors[i]
            dict1 = cv2.pointPolygonTest(pts,(x,y) ,True)
            if dict1 > 0:
                cv2.rectangle(img, (x, y), (x + w, y + h), color, 2)
                cv2.putText(img, label, (x, y + 30), font, 3, color, 3)
            else:
                print(False)
    cv2.imshow("Object detection", img)
    
    key = cv2.waitKey(1)
    if key == ord('q'):
        break

video.release()
```
